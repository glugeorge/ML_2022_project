## Coarse velocities
I am loading in the publically available PISM ensemble results here, to get a sense of what the velocities look like, and to help restrict the forcing within the right time period.

```python
cat = intake.open_catalog('https://raw.githubusercontent.com/ldeo-glaciology/pangeo-pismpaleo/main/paleopism.yaml')
```

I am loading data to dask as it is too big to load otherwise - the following cell needs to be added to match the address of the cluster that is initiated.

```python
cat
```

```python
present  = cat["present"].to_dask()
snapshots = cat["snapshots1ka"].to_dask()
```

Here is what the data structure looks like. We can see how it's a 256 member ensemble based of the 'par' variables

```python
snapshots
```

This is what the average velocity of ice in Antarctica looks like across all members.

```python
mean_vel = snapshots.velbar_mag.mean(dim=['x','y', 'par_esia', 'par_ppq', 'par_prec', 'par_visc']).compute()
mean_vel.plot()
```

## Other forcings
There exist some other forcings that are put into PISM on top of temperature. One notable exclusion in this exploratory data analysis is precipitation. This is because the precipitation forcing used by Albrecht et al. (2020) is a function of temperature, elevation, and the precipitation parameter. Since I am hoping to capture unknown spatial trends (i.e. dependence on elevation) in our models in addition to using temperature and model parameters as inputs, I decided to neglect precipitation. However, there are some other forcings that come out of other model results, so those will be examined.

### Sea level
This data is interesting as it seems to vary a lot slower than temperature. It could potentially be a longer term forcing factor on ice compared to temperature.

```python
sl_forcing = xr.open_dataset('imbrie06peltier15_sl_ramp205ka.nc',decode_times=False)
```

```python
sl_forcing.delta_SL.plot()
```

```python
sl_forcing_interp = sl_forcing.interp(time=snapshots.time)
sl_forcing_interp.delta_SL.plot()
```

### Ocean temperature forcing
This data is interesting as it is 2D and evolving over time. This could have a strong influence over specific ice sheets and marine terminating glaciers.

```python
oc_forcing = xr.open_dataset('schmidtko14_edc-wdc_oceantemp_basins_response_fit_16km-d1.nc',decode_times=False)
```

```python
oc_forcing.theta_ocean.mean(dim='time').plot()
```

```python
oc_forcing.theta_ocean.mean(dim=['x','y']).plot()
```

Overall, these mean forcings for ocean temperature and sea level generally track that of the temperatures (shown in the plot below), so for the sake of simplification, some of my models will only use the temperature forcing, unless 2D spatial data is considered and then I will examine bringing in the ocean forcing. The plot below also highlights the regions selected for training (grey) and testing (green) derived from the data preprocessing notebook. It shows how we are ignoring the deglaciation that occurs starting at around 20 kyr before present as that data lies outside of the training range however we divide it (especially since our testing data has to occur after our training data as it is temporally dependent). 

```python
mean_oc_forcing_interp = oc_forcing.theta_ocean.mean(dim=['x','y']).interp(time=snapshots.time)
ds_forcing_interp = ds_forcing.interp(time=snapshots.time)

fig, axes = plt.subplots(3,1,sharex=True,figsize = (6,8))
mean_vel.plot(ax=axes[0])
ds_forcing_interp.delta_T.plot(ax=axes[1],label='Mean ice surface anomaly')
mean_oc_forcing_interp.plot(ax=axes[1],label='Mean ocean anomaly')
axes[1].legend()
sl_forcing_interp.delta_SL.plot(ax=axes[2])

labels = ['Ice velocity [m/yr]','Temperature Anomaly [K]','Relative Sea Level Anomaly [m]']
for i in range(0,len(axes)):
    axes[i].set_ylabel(labels[i])
    axes[i].set_xlabel('')
    axes[i].axvspan(-1.24e5, -0.55e5, color='grey', alpha=0.3)
    axes[i].axvspan(-0.55e5, -0.35e5, color='green', alpha=0.3)
axes[2].set_xlabel('Time since present (yr)')
```
